```python
df_cat = pd.DataFrame()
df_num = pd.DataFrame()
for y in df.columns:
    if df[y].dtype == object:
        df_cat[y] = df[y]
    else:
        df_num[y] = df[y]
```

### Data Cleaning& Wrangling

#### Finding % missing values using function

```python
def MissingValuePer(df):
    d = {'Column':[],'Missing_Value':[]}
    for i in df.columns:
        d['Column'].append(i)
        d['Missing_Value'].append(str((df[i].isnull().sum()/3425)*100)+'%')
    df_miss = pd.DataFrame(d)
    print(df_miss)
MissingValuePer(df)
```

####  % of missing value Method 2

```python
df_missing = pd.DataFrame()
for y in df.columns:
    mis = (df[y].isnull().sum()/df.shape[0])*100
    data = {'Column':y,'Missing %': mis}
    df_missing = df_missing.append(data,ignore_index=True)
df_missing
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cmap='rainbow')
```

#### Imputation (Dealing with missing values)

```python
def imputation(df,flag):
    if flag == 0:   # fornumerical data
        df.fillna(df.mean(),axis = 0,inplace=True)
    elif flag == 1:   # categorial data
        df.fillna(stats.mode(),axis=0,inplace =True)
imputation(df_num,0)
df_num.isnull().sum()
```

#### Outlier detection & treatment

```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
IQR=q3-q1
lower_bound = q1 - 1.5*IQR
upper_bound = q3 + 1.5*IQR
```

```python
def find_outliers_IQR(df):
    return ((df<lower_bound) | (df>upper_bound)).sum()
find_outliers_IQR(df_num)
```

```python
l = []
def outlier(df):
    for i in df.columns:
        q1=df[i].quantile(0.25)
        q3=df[i].quantile(0.75)
        IQR=q3-q1
        lower_bound = q1 - 1.5*IQR
        upper_bound = q3 + 1.5*IQR
        lst = df.index[(df[i] < lower_bound) | (df[i] > upper_bound)]
        median = df[i].median()
        for index in lst:
            df[i].loc[index] = median
        l.append(lst)
outlier(df_num)
```

### Feature Transformation

```python
d1,d2 = {},{}
def log_transform(df):
    for i in df.columns:
        d1[i],d2[i] = [],[]
        for j in range(3698):
            d1[i].append(np.log(df[i][j]))
            d2[i].append(np.exp(df[i][j]))
log_transform(df)
df_log_values = pd.DataFrame(d1)
df_exp_values = pd.DataFrame(d2)
```

```python
df_log = pd.DataFrame()
df_exp  = pd.DataFrame()
for col in df_num.columns:
    df_log[col] = np.log(df_num[col])
    df_exp[col] = np.exp(df_num[col])
```
